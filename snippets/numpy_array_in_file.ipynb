### Save Array

```python
import numpy as np
arr = np.random.randn(20,10)
np.save('large_arr', arr)
np.savetxt('large_arr.txt', arr) ## savetxt can save only 1d or 2d arr
```

### Load Array

```python
print(np.load('large_arr.npy'))
print(np.loadtxt('large_arr.txt'))
```

### Save Multiple arrays

```python
arr1 = np.random.randn(4,3,3)
arr2 = np.random.rand(1,4,3)
arr3 = np.random.randn(3,2,2)
np.savez('multi_array', a=arr1, b=arr2, jpt=arr3)
```

### Load Multiple arrays

```python
dic = np.load('multi_array.npz')
print('a\n', dic['a'])
print('b\n', dic['b'])
print('jpt\n',dic['jpt'])
```
