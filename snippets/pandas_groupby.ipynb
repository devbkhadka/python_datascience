### Group by function

```python
def group(x):
    buckets = [0,5,15,25,40,55]
    bucket_names = "upto 5, 5 to 15, 15 to 25, 25 to 40, 40 to 55, above 55".split(",")
    for i, (low, high) in enumerate(zip(buckets, buckets[1:])):
        if (x>low) & (x<=high):
            return bucket_names[i]
    return bucket_names[-1]
        
data = tip_data.set_index("total_bill")
data.groupby(group).mean() ## The group function will get index values one at a time
```

### Grouping By Dictonary

```python
map_dict = {
    "Wed": "Workday",
    "Thur": "Workday",
    "Fri": "Workday",
    "Sat": "Weekend",
    "Sun": "Weekend"
}

tip_data.set_index("day").groupby(map_dict).mean() ## Group name will be get by mapping index against the dict
```

### Gropping  By Level
In hirarchical data

```python
sizes = "small,medium,large,x-large,xx-large".split(",")
colors = "red,blue,purple,yellow".split(",")

data_dict = {
    "size": np.random.choice(sizes, 100),
    "color": np.random.choice(colors, 100),
    "sold": np.random.choice(range(30,100),100)
}

sales = pd.DataFrame(data_dict).set_index(["size", "color"])
sales.groupby(level="size").mean()
```

### Custom Aggregate Function

```python
def value_range(arr):
    return arr.max() - arr.min()

grouped = tip_data.groupby("day")
grouped.agg(value_range)
```

### Multiple Functions

```python
grouped = tip_data[["day","total_bill", "tip"]].groupby("day")
grouped.agg(["mean", "count", ("range", value_range)])
```

```python
grouped = tip_data.groupby("day")
grouped.agg({
    "total_bill": ["mean", "max"],
    "tip": "mean",
    "time": "count"
})
```

```python
tip_data.groupby("day", as_index=False).mean() ## resets the index
```

### Apply Function to Group

```python
def top(df):
    # returns top 3 rows
    return df.sort_values(by="total_bill")[-3:]

grouped = tip_data.groupby("day")
grouped.apply(top) ## returns 3 rows with top total_bill
```

```python
## General top function
def top2(df, col="total_bill", count=3):
    return df.sort_values(by=col)[-count:]

grouped = tip_data.groupby("day", group_keys=False) ## exclude group key "day" from result
grouped.apply(top2, col="pct_tip", count=2) ## other parameters to top can be supplied to apply method
```

### Bucket/Quantile Analysis

```python
import numpy as np
import pandas as pd
col1 = np.random.normal(40, 8, 100)
data = {
    "col1": col1,
    "col2": np.random.rand(100) * col1
}

df_data = pd.DataFrame(data)

quartiles = pd.qcut(df_data["col1"], 4)

df_data.groupby(quartiles).describe().stack(0)
```

### Group Specific Fill

```python
df_data2 = df_data.copy()
df_data2.loc[np.random.choice(range(100), 15), "col2"] = np.nan
grouped=df_data2.groupby(quartiles, group_keys=False)

def fill_mean(g):
    return g.fillna(g.mean())

grouped.apply(fill_mean).sort_index()
```
