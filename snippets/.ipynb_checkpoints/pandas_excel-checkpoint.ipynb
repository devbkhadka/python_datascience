### pd.read_excel

```python
import pandas as pd
pd.options.display.max_rows = 10
pd.read_excel('data/excel/example.xlsx', index_col=7)

```

### pd.ExcelFile class
- ExcelFile is faster for multi sheet file

```python
xlsx = pd.ExcelFile('data/excel/bills.xlsx')
pd.read_excel(xlsx, 'Bill-3')
```

### pd.read_excel to read multiple worksheet

```python
pd.read_excel('data/excel/bills.xlsx', sheet_name='Bill-1')
```

### Write to excel file

```python
import sys
arr = [
    list('abcdefghijklmnopqrstuvwxyz'),
    list(range(1,27))
]

df = pd.DataFrame(list(zip(*arr)))
df.to_excel('data/excel/output.xlsx')
```

### Write to multiple sheets

```python
writer = pd.ExcelWriter('data/excel/output_multi_sheet.xlsx')
arr1 = list(zip(*[
    list('abcdefghijklmnopqrstuvwxyz'),
    list(range(1,27))
]))
arr2 = list(zip (
    *[
    list('ABCDEFGHIJKLMNOPQRSTUVWXYZ'),
    list(range(1,27))
]
))
df1 = pd.DataFrame(arr1)
df2 = pd.DataFrame(arr2)

df1.to_excel(writer, 'sheet1')
df2.to_excel(writer, 'sheet2')

writer.save()
```
