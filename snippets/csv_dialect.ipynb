```python
import csv
import pandas as pd

class my_dialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = ','
    quotechar='"'
    quoting = csv.QUOTE_MINIMAL
    
with open('data/customers.csv') as csv_file:
    reader = csv.reader(csv_file, dialect=my_dialect)
    headers = [row for _, row in zip(range(1), reader)][0]
    
    
    rows = [row for row in reader]
    display(headers)
    display(pd.DataFrame(rows, columns=headers))
    
```
