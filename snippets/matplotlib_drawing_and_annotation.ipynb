```python
import matplotlib.pyplot as plt

_, (axis) = plt.subplots()
props = {
    "xlim": [0,100],
    "ylim": [0,100],
}
axis.set(**props)

axis.text(50,50, "Drawing Text\n50,50", family="monospace", fontsize=10)

rect = plt.Rectangle([10,10], 5, 15, color="g", alpha=0.5)
circle = plt.Circle([30,30], 15, color="b", alpha=0.5)
axis.add_patch(rect)
axis.add_patch(circle)
```

### Annotations

```python
import pandas as pd
from datetime import datetime

spx = pd.read_csv("data/spx.csv", index_col=0, parse_dates=True)["SPX"]


fig, (ax) = plt.subplots()

spx.plot(ax=ax, style="-.")
props = {
    "xlim":["1/1/2007", "1/1/2011"],
    "ylim": [600, 1800]
}
ax.set(**props)

crisis_dates = [
    (datetime(2007,10,11), "Peak of bull market"),
    (datetime(2008,3,12), "Bear Starns Fail"),
    (datetime(2008,9,15), "Lehmn Bankrupty")
]

for date, label in crisis_dates:
    value = spx.asof(date)
    arrow_props = dict(facecolor="black", width=2, headwidth=4, headlength=4)
    ax.annotate(label, xy=(date, value+50), xytext=(date, value+100), arrowprops=arrow_props)


```
