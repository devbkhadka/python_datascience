```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

macrodata = pd.read_csv("data/macrodata.csv", index_col=["year", "quarter"])
data = macrodata[["cpi", "m1", "tbilrate", "unemp"]]
changes = np.log(data).diff().dropna()
changes[-5:]
```

```python
sns.regplot("m1", "unemp", data=changes)
plt.title("Relation between change in log ml and log unemp")
```

### Matrix Scatter Plot of Variables

```python
sns.pairplot(changes, diag_kind="kde")
```

### Facet Grid and Categorical Data

```python
tips = pd.read_csv("data/tips.csv")
tips["tip_pct"] = tips["tip"]/(tips["total_bill"]-tips["tip"]) 
tips.head()
```

```python
sns.factorplot(data=tips, x="day", y="tip_pct", col="smoker", row="sex", kind="bar")
```

```python
sns.factorplot(data=tips, x="day", y="tip_pct", col="smoker", kind="box")
```
