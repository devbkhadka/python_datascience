```python
import numpy as np
import pandas as pd

data = pd.DataFrame(np.random.randn(1000,10))

data.describe()
```

### Outliers In A Column

```python
col2 = data[2]
col2[col2.abs()>3]
```

### Outliers in DataFrame

```python
outlier_row_indexes = (np.abs(data)>3).any(1)
outliers = data[outlier_row_indexes]
outliers
```

### Replace Outlier Values

```python
data[np.abs(data)>3] = np.sign(data) * 3
data.describe()
```
