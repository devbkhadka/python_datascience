### Global seed vs Local seed

```python
from numpy import random
random.seed(0)
print(random.rand())
print(random.rand())
```

```python
from numpy import random
rand_state = random.RandomState(0)
print('changes')
print(random.rand())
print(random.rand())
print('doesn\'t change')
print(rand_state.rand())
print(rand_state.rand())
```

### Random Walk
- Start from 0
- In every step move by 1 or -1 with equal probability
- Walk 10000 steps

```python
import matplotlib.pyplot as plt
random_steps = np.where(np.random.choice([0,1],1000)==0,-1,1)
distance_walked = np.cumsum(random_steps)
plt.plot(distance_walked)
plt.show()

print("min distance walked", distance_walked.min())
print("max distance walked", distance_walked.max())

### step when random walk cross distance 10
### first convert distance_walked to boolean if distance is greater than 10 then find index of 1st True 
### (True is max in boolean array)
print((np.abs(distance_walked)>=10).argmax())
```

### 5000 Random Walks Simultaniously

```python
random_steps = np.where(np.random.randn(5000, 1000)>0,1,-1)
distance_walked = np.cumsum(random_steps, 1)
plt.plot(distance_walked[50])
plt.show()
## Find minimum steps to cross distance 30 or -30
target = 20
is_crossing_target = (np.abs(distance_walked)>target).any(1)
crossing_points = (np.abs(distance_walked)>target)[is_crossing_target].argmax(1)

print('Minimum steps to cross target',crossing_points.min())
print('Average distance to cross target', np.mean(crossing_points))
```
