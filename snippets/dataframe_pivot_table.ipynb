```python
import pandas as pd
tips_data = pd.read_csv("data/tips.csv")
tips_data.head()
```

```python
tips_data.pivot_table(index=["day", "time"]) ## aggregates using mean
```

```python
tips_data.pivot_table(["tip", "size"], index=["day", "time"]) ## agrigates only tip and size
```

```python
tips_data.pivot_table(["tip", "size"], index=["day", "time"], columns=["smoker"]) ## group by smoker column
```

```python
tips_data.pivot_table(["tip", "size"], index=["day", "time"], columns=["smoker", "sex"])
```

### Crosstab

```python
pd.crosstab(tips_data.day, tips_data.time) ## counts frequency for each rows and columns combination
```

```python
pd.crosstab([tips_data.day, tips_data.smoker], tips_data.time)
```
